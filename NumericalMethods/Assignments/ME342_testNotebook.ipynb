Jupyter notebooks are similar in appearance to Mathematica notebooks.  To evaluate a cell, place the curser on the cell and push the Shift and Enter keys at the same time and in that order.  (In other words, push the enter key while holding the shift key, then release.)

There are three types of cells in a Jupyter notebook.  <em>Code</em> cells are for executable code (e.g., Python scripting).  <em>Raw NBConvert</em> cells will simply cause the contents to be ignored when executed.  Simply change a cell to this type when you want an entire cell to be commented out of execution.  The third type is a <em>Markdown</em> cell.  Markdown cells are for media content; for instance, this text is contained in a Markdown cell.  "Markdown" is actually a markup language, like HTML.  A Markdown cell can render Markdown, HTML, LaTeX, and other languages or graphics simultaneously.  After highlighting/selecting this cell, press the enter key to see the Markdown, HTML, and LaTeX code embedded within it.  Evaluate the cell (Shift + Enter) to render the contents again.

Cells can be created, manipulated/moved, and deleted using shortcut keys.  Line numbers can also be displayed in a code cell (useful for tracing error-message references).  Click on the Help menu --> Keyboard Shortcuts above to see a list of possibilities.  

Jupyter can use different language kernels; Python is usually the default one.  The circle to the right of the Python kernel indicator (upper right of page) will be filled when the interpreter is busy (evaluating something) and hollow when it is idle.  You can interupt the interpreter with the square button above.  Like Mathematica and Matlab, the kernel/interpreter retains the state of previous evaluations while running in linkage with a Jupyter notebook.  You can restart a kernel to clear all variables and imported libraries; doing so prior to re-evaluating/checking your notebook is <em>highly</em> recommended before you submit it as a homework assignment.  The "Run All" option under the "Cell" menu above will evaluate your entire notebook from the top down.  This option will commonly be employed when your assignment is graded; so please ensure prior to submission that your notebook behaves as you want it to when this option is used.
